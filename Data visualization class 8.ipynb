```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
i_94 = pd.read_csv(r"C:\Users\Dell\Metro Interstate Traffic Volume\Metro_Interstate_Traffic_Volume.csv")
```

```python
i_94.head()
```

```python
i_94.date_time = pd.to_datetime(i_94.date_time)
```

```python
i_94.info()
```

```python
day_bol = (i_94.date_time.dt.hour >= 7) & (i_94.date_time.dt.hour < 19)
day = i_94[day_bol].copy()
day
```

```python
day["dayofweek"] = day.date_time.dt.dayofweek
day
```

```python
day["dayofweek"].unique()
```

```python
day["dayinwords"] = day.date_time.dt.strftime("%A")
day
```

```python
day[day["dayinwords"]== "Monday"]
```

```python
import datetime as dt
```

```python
avg_dayofweek = day.groupby("dayofweek")["traffic_volume"].mean()
avg_dayofweek
```

```python
avg_dayofweek.plot.line()
plt.show()
```

```python
avg_dayofweek.plot(kind = "line")
plt.show()
```

```python
day["hour"] = day.date_time.dt.hour
day
```

```python
weekdays = day[day["dayofweek"] <= 4]
weekdays
```

```python
weekend = day[~ (day["dayofweek"] <= 4)]      # ~ means (not this)
weekend
```

```python
weekdays.pivot_table(index = "hour", values = "traffic_volume",)
```

```python
plt.figure(figsize = (11,4))
plt.subplot(1,2,1)

hourly_weekday_avg = weekdays.pivot_table(index = "hour", values="traffic_volume")
plt.plot(hourly_weekday_avg.index, hourly_weekday_avg)
plt.ylim(1500,6500)

plt.subplot(1,2,2)

hourly_weekend_avg = weekend.pivot_table(index = "hour", values="traffic_volume")
plt.plot(hourly_weekend_avg.index, hourly_weekend_avg)
plt.ylim(1500,6500)

plt.show()
```
